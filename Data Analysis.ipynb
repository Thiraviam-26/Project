## Encoding

```python
churnNN=churn.dropna(axis=1)
churnNN.head()
```

```python
churnNN.info()
```

```python
churnNN.drop('CustID',axis=1,inplace=True)
```

```python
tobeEnc=churnNN.select_dtypes(include=['object']).columns
tobeEnc
```

```python
from sklearn.preprocessing import OneHotEncoder
import sklearn
sklearn.__version__
#0.20.0#
```

```python
enc = OneHotEncoder(handle_unknown='ignore')
enc.fit(churnNN[tobeEnc])
```

```python
enc.categories_
```

```python
churnEncoded=enc.transform(churnNN[tobeEnc])
churnEncoded
```

```python
churnEncDF=pd.DataFrame(churnEncoded[:10].toarray(),columns=enc.get_feature_names())
churnEncDF
```

```python
churnNN.head()
```

```python
churnNN.drop(tobeEnc,axis=1,inplace=True)
churnFinalEnc=pd.concat([churnNN,churnEncDF],axis=1)
churnFinalEnc.head()
```

## Model Creation

```python
bike.drop('datetime',axis=1,inplace=True)
```

```python
bike.head()
```

```python
from sklearn.model_selection import train_test_split

X=bike.drop('count',axis=1).values
y=bike['count']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor

rf=RandomForestRegressor()
rf.fit(X_train,y_train)
```

```python
rf.feature_importances_
```

```python
rf.predict(X_test)
```

```python
predictions=rf.predict(X_test)
predComp=pd.DataFrame({'Predictions':predictions,'Actual':y_test})
predComp.head()
```

```python
a4_dims = (20, 10)
fig, ax = pyplot.subplots(figsize=a4_dims)

sns.lineplot(ax=ax,data=predComp,palette='tab10')
```

```python
predComp.to_csv("finalPred.csv")
```
